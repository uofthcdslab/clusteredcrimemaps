```python
import os 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
frame = pd.read_csv('all_data', sep=';', index_col=False)
```

```python
error = frame.isnull().sum(1).sort_values(ascending=False)
error = error[error != 0]
#frame = frame.drop(error.index)
```

```python
frame.sort_values(['name','k'])
index = frame.columns[2:]
frame['Average'] = frame.loc[:,index].mean(1)
```

```python

for k in range(5,6):
    sub_frame = frame[frame.k==k]
#     sub_frame = frame[frame.name.str.contains('2015')]
#     sub_frame = sub_frame[sub_frame.k==k]
    freq_list = []
    for i in sub_frame.index:
        freq_list += sub_frame.loc[i,index].tolist()
    plt.figure(k)
    plot = sns.distplot(freq_list, kde=False)
    plot.set(xlabel='Potential Bias Index',ylabel='Number of Occurences')
    sns.plt.title(str(k) + ' Clusters')
```
