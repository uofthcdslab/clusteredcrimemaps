```python
import pandas as pd
import numpy as np
import geopandas as gpd
from sklearn.cluster import KMeans

from shapely.geometry import Point
from geopandas import GeoDataFrame
from bokeh.models import HoverTool
```

```python
df = pd.read_csv("../CrimeData/TheftData.csv", sep =';')

x = []
y = []
def gen_coords(loc):
    data = loc[1:-1].split(',')
    data = list((float(data[0]), float(data[1])))
    x.append(data[1])
    y.append(data[0])

    
df['Location'].apply(gen_coords)
points = [Point(xy) for xy in zip(x,y)]
crs = {'init': 'epsg:4326'}
geo_df = GeoDataFrame(df,crs=crs, geometry=points)
m_theft = geo_df[geo_df['Offense 1'] == 'MOTOR VEHICLE THEFT']
theft = geo_df.copy()

for index, crime in m_theft['Offense 1'].items():
    theft = theft.drop(index)
    
    
t_list = []
mt_list = []

for index, point in theft['geometry'].items():
    t_list.append([point.coords.xy[0][0], point.coords.xy[1][0]])
    
for index, point in m_theft['geometry'].items():
    mt_list.append([point.coords.xy[0][0], point.coords.xy[1][0]])
```

```python
X = np.array(t_list)
Y = np.array(mt_list)

for k in range(2, 11):
    for x in range(10):
        kmeans_theft = KMeans(n_clusters = k, init = 'random', n_init = 1).fit(X)
        kmeans_motor = KMeans(n_clusters = k, init = 'random', n_init = 1).fit(Y)
        index = 'K' + str(k) + "-" + str(x)
        theft.loc[:,index] =  kmeans_theft.labels_
        m_theft.loc[:,index] = kmeans_motor.labels_


```

```python
theft.to_file('tPoints.js', driver='GeoJSON')
m_theft.to_file('mPoints.js', driver='GeoJSON')
```

```python
shape_filepath = r"../CrimeData/neighborhood/neighborhood.shp"

shapes = gpd.read_file(shape_filepath)
shapes['geometry'] = shapes['geometry'].to_crs(epsg=4326)

def count_points(polygon, points):
    i = 0
    for index, point in points:
        if polygon.contains(point):
            i += 1
    return i
        

t_count=[]
m_count=[]
for index, polygon in shapes['geometry'].items():
    t_count.append(count_points(polygon, theft['geometry'].items()))
    m_count.append(count_points(polygon, m_theft['geometry'].items()))
```

```python
shapes['T Count'] = t_count.copy()
shapes['M Count'] = m_count.copy()
```

```python
shapes.to_file('cityLimit.js', driver = 'GeoJSON')
```
